# Hello PyTorch Image Preprocessing  

```python
# Prerequisites
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

### Open Image

```python
img = Image.open('images/motorcycle.jpg')
img
```

```python
img.size
```

### Compose Processing Steps

```python
preprocess_steps = transforms.Compose([
    transforms.Resize((512, 512)),  
    transforms.RandomRotation(50),
    transforms.CenterCrop(300),
    transforms.Grayscale(),
    transforms.RandomVerticalFlip(),
])
x = preprocess_steps(img)
x 
```

```python
preprocess_steps = transforms.Compose([
    transforms.Resize((512, 512)),  
    transforms.RandomRotation(50),
    transforms.CenterCrop(400),
    transforms.Grayscale(),
    transforms.RandomVerticalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5]),  # For grayscale
    #transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),  # ImageNet values
])
x = preprocess_steps(img)
print("x shape: ", x.shape)
x
```

```python
print("x mean:", torch.mean(x))
print("x std:", torch.std(x))
```
